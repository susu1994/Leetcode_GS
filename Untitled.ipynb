```python
import matplotlib.pyplot as plt
import chaospy as cp
import numpy as np
def LGC(a,c, m:int, z):
    output = []
    for i in range(m):
        mod = (a*z+c)%m
        output.append(mod/m)
        z = mod
    return output
a=6.5
c=0.7
m=60
z=2.23
output = LGC(a,c,m,z)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(np.arange(1,21,1), output[0:20], color='r')
ax.scatter(np.arange(21,41,1), output[20:40], color='b')
ax.scatter(np.arange(41,61,1), output[40:60], color='g')
# dimension = 2
# sequence_count = 200

# # 生成Sobol序列
# sobol_samples = cp.sampler.generate_samples(
#     (sequence_count, dimension), rule="sobol"
# )

# # 打印生成的Sobol序列
# print(sobol_samples)
```
