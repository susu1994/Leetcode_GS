```python
from typing import List
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.

        """
        leng = len(matrix[0])
        for i in range(leng//2):
            for j in range(i,leng-i-1):
                temp=matrix[j][leng-1-i]
                matrix[j][leng-1-i]=matrix[i][j]
                matrix[i][j] =  matrix[leng-1-j][i]
                matrix[leng-1-j][i]=matrix[leng-1-i][leng-1-j]
                matrix[leng-1-i][leng-1-j]=temp
test = Solution()
abc = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]     
test.rotate(abc)
print(abc)
```

```python
class Solution:
    def rotate1(self, matrix: List[List[int]]) -> None:
        self.transpose(matrix)
        self.reflect(matrix)
    
    def transpose(self, matrix):
        n = len(matrix)
        for i in range(n):
            for j in range(i + 1, n):
                matrix[j][i], matrix[i][j] = matrix[i][j], matrix[j][i]

    def reflect(self, matrix):
        n = len(matrix)
        for i in range(n):
            for j in range(n // 2):
                matrix[i][j], matrix[i][-j - 1] = matrix[i][-j - 1], matrix[i][j]
test2 = Solution()
abc = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]     
test2.rotate1(abc)
print(abc)
```
