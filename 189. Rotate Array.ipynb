```python
from typing import List
import copy
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        rot = k%len(nums)
        first = nums[len(nums)-rot:]
        second= nums[:len(nums)-rot]
        nums[:] = first+second

test = Solution()
a = [1,2,3,4,5,6,7]
test.rotate(a,3)
print(a)
```
