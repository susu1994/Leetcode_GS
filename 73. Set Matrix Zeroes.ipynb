```python
from typing import List
class Solution:
       def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        ind = []
        n=len(matrix)
        m=len(matrix[0])
        for row in range(n):
            for col in range (m):
                if matrix[row][col]==0:
                    ind.append([row,col])
        for a in ind:
            matrix[a[0]]=[0]*m
            for i in range(n):
                 matrix[i][a[1]]=0
        print(matrix)
test = Solution()
test.setZeroes([[0,1,2,0],[3,4,5,2],[1,3,1,5]])
```
